```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,50) # assignment doesn't explicitly say it but I'm assuming this is to be x
y = np.log(x)
z = np.exp(x)
```

```python
plt.scatter(x,y, label = 'Natural Log')
plt.scatter(x,z, label = 'Exponential')

#adding in labels for pt 6-7
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Natural Log and Exponential Functions')
plt.legend(loc = 'lower right')

plt.show()
```

```python
varx = np.linspace(-6.5,6.5,100)
vary = np.sin(varx)
varz = np.cos(varx)
```

```python
plt.scatter(varx,vary)
plt.scatter(varx,varz)
plt.show()
```

```python
plt.plot(varx,vary)
plt.plot(varx,varz)
plt.show()
```

```python
plt.plot(varx,vary,label='sin')
plt.plot(varx,varz, label ='cos')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sin and Cos Functions')
plt.legend(loc = 'lower right')
plt.show()
```
